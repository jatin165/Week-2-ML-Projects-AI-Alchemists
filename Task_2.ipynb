```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
print("Dataset:")
print(df.head())
```

```python
data = df[['reading score', 'writing score']].head(5)
```

```python
X = data['reading score'].values
Y = data['writing score'].values
```

# Step 1: Calculate Means

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

print("Mean of Reading Score:", mean_x)
print("Mean of Writing Score:", mean_y)
```

Step 2: Calculate Slope

```python
numerator = np.sum((X - mean_x) * (Y - mean_y))
denominator = np.sum((X - mean_x) ** 2)

m = numerator / denominator

print("Slope (m):", m)
```

Step 3: Calculate Intercept

```python
c = mean_y - (m * mean_x)

print("Intercept (c):", c)
```

Step 4: Predict Values

```python
Y_pred = m * X + c

print("First 5 Predicted Values:")
print(Y_pred[:5])
```

Step 5: Calculate Mean Squared Error

```python
mse = np.mean((Y - Y_pred) ** 2)

print("Mean Squared Error:", mse)
```

Step 6: Compare Actual and Predicted Values

```python
result = pd.DataFrame({
    "Reading Score": X[:5],
    "Actual Writing Score": Y[:5],
    "Predicted Writing Score": np.round(Y_pred[:5], 2)
})

print(result)
```

Step 7: Visualization

```python
plt.figure(figsize=(8,5))

plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')

plt.title("Linear Regression from Scratch")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.legend()

plt.show()
```
